### Estimating the periodicity

Periodical cicadas live in groups called broods. Cicada broods have either a 17-year or 13-year lifecycle. A brood naming system was established in 1898 using Roman numeral names. Here are the largest broods for each lifecycle duration:

| Lifecycle Duration | Brood Names      |
|--------------------|------------------|
| 13 years           | XIX, XXII, XXIII |
| 17 years           | II, IV, X, XIV   |

The media will often announce ahead of time when different broods are about to emerge from underground. Can you estimate which lifecycle gets the most attention according to our word usage data?


Activity Goal:
- Use the ruler to estimate the period duration.

```python
plot_animal('cicada')
plt.xlim(1820, 2020)
plt.axvline(x=1898, color='magenta', linestyle='--', label='1898 Naming System')
plt.legend(loc='lower right')
plt.show()
```

Is the periodicity closer to a 13-year or 17-year cycle?

### Which 17-year brood?
It looks like 17-year cicadas receive the most attention. The four largest 17-year broods are: X, XIV, II, and IV.

We have historical records from 1915 onward for when each of these broods emerged:

| Brood   | Years Emerged             |
|---------|---------------------------|
| Brood X | 1919, 1936, 1953, 1970, 1987, 2004 |
| Brood XIV | 1923, 1940, 1957, 1974, 1991, 2008 |
| Brood II | 1928, 1945, 1962, 1979, 1996, 2013 |
| Brood IV | 1930, 1947, 1964, 1981, 1998, 2015 |

Can we use this data to infer which of these 17-year broods influenced word usage the most? Let's focus on the years from 1915 to 1970, where distinct cyclic peaks and valleys are evident.

Activity Goal:

```python
# helper function
def add_emergence_years( brood_info={} ):
	title = brood_info['title']
	years = brood_info['years']
	for year in years:
		plt.axvline( year, color='magenta', ls=':' )
	plt.title(f'{title}: Years Emerged')

# brood info
brood_x = {'years': [1919, 1936, 1953], 'title':'Brood X'}
brood_xiv = {'years': [1923,1940,1957], 'title':'Brood XIV'}
brood_ii = {'years': [1928,1945,1962], 'title':'Brood II'}
brood_iv = {'years': [1930,1947,1964], 'title':'Brood IV'}
```

```python
plot_animal('cicada')
plt.xlim(1820, 2020)
add_emergence_years( brood_x )
# try brood_x, brood_xiv, brood_ii, brood_iv
```

**Investigating the Multi-Year Repeating Patterns of Cicadas Using Word Frequency and Print Mentions**


This study investigates the cyclical emergence patterns of periodical cicadas through linguistic data analysis. By examining the historical frequency of the word “cicada” in print and comparing it to known cicada brood emergence years, I test the hypothesis that media attention peaks align with specific cicada broods. My results support a 17-year cycle, with Brood II receiving the most consistent and significant media attention.


Periodical cicadas are a unique phenomenon in the natural world, emerging in large groups every 13 or 17 years after spending most of their lives underground. The sudden and synchronised appearance of these insects often generates public interest and media coverage. This project explores whether such biological cycles are reflected in word usage data and how they correspond to the emergence years of different cicada broods.


I used two datasets:

1. `animal-word-trends-cicada.csv` – Contains annual word frequency data (per million words) specifically for the word "cicada".

2. `cicada-brood-mentions-print.csv` – Reports the total number of print mentions for cicada broods between 1915 and 1970.

I used a Python-based plotting function plot_animal() to visualise trends in the word "cicada" from 1820 to 2020. Vertical lines were added to highlight known emergence years of 17-year broods (X, XIV, II, and IV) to identify alignment with word frequency peaks.


```python
import pandas as pd
import matplotlib.pyplot as plt

#  dataset
word_trends = pd.read_csv('animal-word-trends-cicada.csv')

#  plotting function
def plot_animal(animal):
    trend = word_trends.query("word == @animal")
    plt.plot(trend['year'], trend['frequency'], label=animal)
    plt.ylabel('Frequency per million')

# Plot for 'cicada'
plot_animal('cicada')
plt.xlim(1820, 2020)
plt.title('Word Usage Trend for "Cicada"')
plt.show()
```

**Findings**

1. **Periodic Word Usage**

   Visual inspection of the graph for the word “cicada” revealed repeating peaks approximately every 17 years. These peaks suggest increased media attention during periods of cicada emergence.

2. **Alignment with Broods**

   I compared the emergence years of major 17-year broods to the peaks in the graph. Brood II emergence years (1928, 1945, 1962) aligned strongly with the highest peaks in word frequency data. Brood X emergence years (1919, 1936, 1953) also aligned with smaller peaks.

```python
# vertical lines for Brood II years
plot_animal('cicada')
plt.xlim(1820, 2020)
plt.title('Brood II: Years Emerged')
for year in [1928, 1945, 1962]:
    plt.axvline(year, color='magenta', linestyle=':')
plt.show()
```

3. **Print Mention Analysis**

   The bar chart of print mentions from 1915–1970 showed that Brood II had the highest number of mentions (1024), followed by Brood X (968). Other broods such as XIV and IV had significantly fewer mentions, reinforcing the hypothesis that Brood II garners the most attention.

```python
broods = pd.read_csv('cicada-brood-mentions-print.csv')

plt.bar(broods['brood_name'], broods['mentions'])
plt.xlabel('Brood Name')
plt.ylabel('Number of Mentions')
plt.title('Total Mentions in Print for Each Cicada Brood')
plt.show()
```


The 17-year life cycle of periodical cicadas is strongly reflected in language usage patterns. Peaks in the word “cicada” correlate with the emergence of Brood II, suggesting it receives the most sustained media interest. Brood X also receives notable attention, but to a slightly lesser extent. The consistency between linguistic data and biological emergence years confirms that public discourse mirrors cicada lifecycle patterns.

This research supports the hypothesis that periodical cicada emergence, particularly from Brood II, drives increased word usage and media attention every 17 years. Such findings highlight the intersection of biological events and cultural response, revealing how natural phenomena are documented and amplified in human language.
